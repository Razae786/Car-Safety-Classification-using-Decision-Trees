```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

#As their is no header in my dataset so I am creating this
df = pd.read_csv('car.csv', header=None, names=[
    'buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class'
])
df.head()
```

```python
df.isnull().sum()
```

```python
# Separate features and target
inputs = df.drop('class', axis='columns')
target = df['class']
df.head()
```

```python
# Encode all categorical columns
le_buying = LabelEncoder()
le_maint = LabelEncoder()
le_doors = LabelEncoder()
le_persons = LabelEncoder()
le_lug_boot = LabelEncoder()
le_safety = LabelEncoder()
le_class = LabelEncoder()

inputs['buying_n'] = le_buying.fit_transform(inputs['buying'])
inputs['maint_n'] = le_maint.fit_transform(inputs['maint'])
inputs['doors_n'] = le_doors.fit_transform(inputs['doors'])
inputs['persons_n'] = le_persons.fit_transform(inputs['persons'])
inputs['lug_boot_n'] = le_lug_boot.fit_transform(inputs['lug_boot'])
inputs['safety_n'] = le_safety.fit_transform(inputs['safety'])

# After encording
inputs.head()
```

```python
# Drop original columns
inputs_n = inputs.drop(['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety'], axis='columns')
inputs_n.head()
```

```python
# Encode the target column
target_n = le_class.fit_transform(target)
target_n
```

```python
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(inputs_n, target_n, test_size=0.2, random_state=42)
# Displaying this
print("Training set size:", len(X_train))
print("Test set size:", len(X_test))
```

```python
# Train Decision Tree model
model = tree.DecisionTreeClassifier()
model.fit(X_train, y_train)
print("Training Accuracy:", model.score(X_train, y_train))
print("Test Accuracy:", model.score(X_test, y_test))
```

```python
y_pred = model.predict(X_test)
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=le_class.classes_))
```

```python
# Sample Example
sample = [
    le_buying.transform(['vhigh'])[0],
    le_maint.transform(['vhigh'])[0],
    le_doors.transform(['2'])[0],
    le_persons.transform(['2'])[0],
    le_lug_boot.transform(['small'])[0],
    le_safety.transform(['low'])[0]
]

sample_df = pd.DataFrame([sample], columns=[
    'buying_n', 'maint_n', 'doors_n', 'persons_n', 'lug_boot_n', 'safety_n'
])

# Predict
prediction = model.predict(sample_df)
print("The predicted Result is:")
print("Predicted Class for sample:[", le_class.inverse_transform(prediction)[0],"]")
```
